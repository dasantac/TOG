# Feature transformations

## Active Hand data

### Hand transformations

#### Center of Gravity

```python
PH1_AH_cog_df = PH1_active_df.copy()

PH1_AH_cog_df[ph2.pf_mean_hand_columns] = PH1_AH_cog_df[ph2.pf_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)
PH1_AH_cog_df
```

#### Normal Vector

```python
PH1_AH_v123_df = PH1_AH_cog_df.copy()

PH1_AH_v123_df[ph2.pf_v123_columns] = PH1_AH_v123_df[ph2.handedness_column+ph2.pf_hand_landmark_columns].apply(ph2.hand.NormalVector, axis=1)
PH1_AH_v123_df
```

#### Change of Base

```python
PH1_AH_wrist_df = PH1_AH_v123_df.copy()

PH1_AH_wrist_df[ph2.pf_wrist_hand_landmark_columns] = PH1_AH_wrist_df[ph2.pf_hand_landmark_columns+ph2.pf_v123_columns].apply(ph2.hand.ChangeOfBase, axis=1)
PH1_AH_wrist_df
```

## Per Video data

### Hand transformations

#### Center of Gravity

```python
PH1_VID_cog_df = PH1_video_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = ph2.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_mean_hand_columns = ph2.pv_mean_hand_columns[3*frame:3*(frame+1)]
  PH1_VID_cog_df[current_frame_mean_hand_columns] = PH1_VID_cog_df[current_frame_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)

PH1_VID_cog_df
```

#### Normal Vector

```python
PH1_VID_v123_df = PH1_VID_cog_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = ph2.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_v123_columns = ph2.pv_v123_columns[9*frame:9*(frame+1)]
  PH1_VID_v123_df[current_frame_v123_columns] = PH1_VID_v123_df[ph2.handedness_column+current_frame_hand_landmark_columns].apply(ph2.hand.NormalVector, axis=1)

PH1_VID_v123_df
```

#### Change of Base

```python
PH1_VID_wrist_df = PH1_VID_v123_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = ph2.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_v123_columns = ph2.pv_v123_columns[9*frame:9*(frame+1)]
  current_frame_wrist_hand_landmark_columns = ph2.pv_wrist_hand_landmark_columns[63*frame:63*(frame+1)]
  PH1_VID_wrist_df[current_frame_wrist_hand_landmark_columns] = PH1_VID_wrist_df[current_frame_hand_landmark_columns+current_frame_v123_columns].apply(ph2.hand.ChangeOfBase, axis=1)

PH1_VID_wrist_df
```
