# Feature transformations

## Per frame data

### Hand transformations

#### Center of Gravity

```python
PH2_FRM_h_cog_df = PH1_frame_df.copy()

PH2_FRM_h_cog_df[sup.pf_mean_hand_columns] = PH2_FRM_h_cog_df[sup.pf_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)
PH2_FRM_h_cog_df
```

#### Normal Vector

```python
PH2_FRM_h_v123_df = PH2_FRM_h_cog_df.copy()

PH2_FRM_h_v123_df[sup.pf_h_v123_columns] = PH2_FRM_h_v123_df[[sup.handedness_column]+sup.pf_hand_landmark_columns].apply(ph2.hand.NormalVector, axis=1)
PH2_FRM_h_v123_df
```

#### Change of Base

```python
PH2_FRM_h_wrist_df = PH2_FRM_h_v123_df.copy()

PH2_FRM_h_wrist_df[sup.pf_wrist_hand_landmark_columns] = PH2_FRM_h_wrist_df[sup.pf_hand_landmark_columns+sup.pf_h_v123_columns].apply(ph2.hand.ChangeOfBase, axis=1)
PH2_FRM_h_wrist_df
```

### Pose transformations

#### Normal Vector

```python
PH2_FRM_p_v123_df = PH2_FRM_h_wrist_df.copy()

PH2_FRM_p_v123_df[sup.pf_p_v123_columns] = PH2_FRM_p_v123_df[sup.pf_pose_landmark_columns].apply(ph2.pose.NormalVector, axis=1)
PH2_FRM_p_v123_df
```

#### Change of Base

```python
PH2_FRM_p_chest_df = PH2_FRM_p_v123_df.copy()

PH2_FRM_p_chest_df[sup.pf_chest_pose_landmark_columns] = PH2_FRM_p_chest_df[sup.pf_pose_landmark_columns+sup.pf_p_v123_columns+sup.pf_mean_hand_columns].apply(ph2.pose.ChangeOfBase, axis=1)
PH2_FRM_p_chest_df
```

## Per Video data

### Hand transformations

#### Center of Gravity

```python
PH2_VID_h_cog_df = PH1_video_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = sup.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_mean_hand_columns = sup.pv_mean_hand_columns[3*frame:3*(frame+1)]
  PH2_VID_h_cog_df[current_frame_mean_hand_columns] = PH2_VID_h_cog_df[current_frame_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)

PH2_VID_h_cog_df
```

#### Normal Vector

```python
PH2_VID_h_v123_df = PH2_VID_h_cog_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = sup.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_v123_columns = sup.pv_h_v123_columns[9*frame:9*(frame+1)]
  PH2_VID_h_v123_df[current_frame_v123_columns] = PH2_VID_h_v123_df[[sup.handedness_column]+current_frame_hand_landmark_columns].apply(ph2.hand.NormalVector, axis=1)

PH2_VID_h_v123_df
```

#### Change of Base

```python
PH2_VID_h_wrist_df = PH2_VID_h_v123_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = sup.pv_hand_landmark_columns[63*frame:63*(frame+1)]
  current_frame_v123_columns = sup.pv_h_v123_columns[9*frame:9*(frame+1)]
  current_frame_wrist_hand_landmark_columns = sup.pv_wrist_hand_landmark_columns[63*frame:63*(frame+1)]
  PH2_VID_h_wrist_df[current_frame_wrist_hand_landmark_columns] = PH2_VID_h_wrist_df[current_frame_hand_landmark_columns+current_frame_v123_columns].apply(ph2.hand.ChangeOfBase, axis=1)

PH2_VID_h_wrist_df
```

### Pose transformations

#### Normal Vector

```python
PH2_VID_p_v123_df = PH2_VID_h_wrist_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_pose_landmark_columns = sup.pv_pose_landmark_columns[9*frame:9*(frame+1)]
  current_frame_v123_columns = sup.pv_p_v123_columns[9*frame:9*(frame+1)]
  PH2_VID_p_v123_df[current_frame_v123_columns] = PH2_VID_p_v123_df[current_frame_pose_landmark_columns].apply(ph2.pose.NormalVector, axis=1)

PH2_VID_p_v123_df
```

#### Change of Base

```python
PH2_VID_chest_df = PH2_VID_p_v123_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_pose_landmark_columns = sup.pv_pose_landmark_columns[9*frame:9*(frame+1)]
  current_frame_v123_columns = sup.pv_p_v123_columns[9*frame:9*(frame+1)]
  current_frame_mean_hand_columns = sup.pv_mean_hand_columns[3*frame:3*(frame+1)]
  current_frame_chest_pose_landmark_columns = sup.pv_chest_pose_landmark_columns[12*frame:12*(frame+1)]
  PH2_VID_chest_df[current_frame_chest_pose_landmark_columns] = PH2_VID_chest_df[current_frame_pose_landmark_columns+current_frame_v123_columns+current_frame_mean_hand_columns].apply(ph2.pose.ChangeOfBase, axis=1)

PH2_VID_chest_df
```
