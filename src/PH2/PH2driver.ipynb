# Feature transformations

## Active Hand data

### Hand transformations

#### Center of Gravity

```python
PH1_AH_cog_df = PH1_active_df.copy()

PH1_AH_cog_df[["h_mean_x", "h_mean_y", "h_mean_z"]] = PH1_AH_cog_df[ph2.frame_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)
PH1_AH_cog_df
```

## Per Video data

### Hand transformations

#### Center of Gravity

```python
PH1_VID_cog_df = PH1_video_df.copy()

for frame in range(sup.NUM_FRAMES_PER_VIDEO):
  current_frame_hand_landmark_columns = ph2.video_hand_landmark_columns[12*frame:12*(frame+1)]
  PH1_VID_cog_df[[f"f_{frame}_h_mean_x", f"f_{frame}_h_mean_y", f"f_{frame}_h_mean_z"]] = PH1_VID_cog_df[current_frame_hand_landmark_columns].apply(ph2.hand.CenterOfGravity, axis=1)

PH1_VID_cog_df
```
